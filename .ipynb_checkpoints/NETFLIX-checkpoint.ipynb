```python
x=df.drop('Close',axis=1)
y=df['Close']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
print(X_TRAIN.shape)
print(X_TEST.shape)
print(Y_TRAIN.shape)
print(Y_TEST.shape)
```

Normalization data using standardscalr

```python
from sklearn.preprocessing import StandardScaler
```

```python
sclr=StandardScaler()
```

```python
X_TRAIN=sclr.fit_transform(X_TRAIN)
X_TEST=sclr.transform(X_TEST)
```

```python
X_TRAIN
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_TRAIN,Y_TRAIN)
```

```python
y_pred=lr.predict(X_TEST)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(Y_TEST,y_pred)
```

```python
mean_squared_error(Y_TEST,y_pred)
```

Prediction model

```python
df.columns
```

```python
def pred(Open, High, Low, Adj_Close, Volume, year, month,
       day):
  features= np.array([[Open, High, Low, Adj_Close, Volume, year, month,
       day]])
  features=sclr.fit_transform(features)
  prediction = lr.predict(features).reshape(1,-1)
  return prediction[0]
```

```python
Open= 1.214286
High= 1.225000
Low = 1.197143
Adj_Close= 1.210000
Volume= 11104800
year= 2002
month= 5
day= 24

res=pred(Open, High, Low, Adj_Close, Volume, year, month,
       day)
```
